```python
# Import dependencies.
import pandas as pd
```

```python
# Load housing csv.
file1 = '../Data/Clean_Housing.csv'
housing_df = pd.read_csv(file1)

print(housing_df.shape)
housing_df.head()
```

```python
# Load weather csv.
file2 = '../Data/Clean_Weather.csv'
weather_df = pd.read_csv(file2)

print(weather_df.shape)
weather_df.head()
```

```python
# Merge housing and weather DataFrames.
housing_weather_df = housing_df.merge(weather_df, on='City', how='left')

print(housing_weather_df.shape)
housing_weather_df.head()
```

```python
# Check for null values.
housing_weather_df.isnull().sum()
```

```python
# Drop null values.
housing_weather_df = housing_weather_df.dropna()
```

```python
# Validate null values dropped.
housing_weather_df.isnull().sum()
```

```python
# Load census csv.
file3 = '../Data/Clean_Census.csv'
census_df = pd.read_csv(file3)

print(census_df.shape)
census_df.head()
```

```python
# Load population csv.
file4 = '../Data/Clean_Population.csv'
population_df = pd.read_csv(file4)

print(population_df.shape)
population_df.head()
```

```python
# Drop population.
population_df = population_df.drop(columns='Population')
```

```python
# Merge census and population DataFrames.
population_census_df = population_df.merge(census_df, on='County', how='left')

print(population_census_df.shape)
population_census_df.head()
```

```python
# Check for null values.
population_census_df.isnull().sum()
```

```python
# Drop null values.
population_census_df = population_census_df.dropna()
```

```python
# Validate null values.
population_census_df.isnull().sum()
```

```python
population_census_df.shape
```

```python
# Final merge.
merged_df = housing_weather_df.merge(population_census_df, on='City', how='left')

print(merged_df.shape)
merged_df.head()
```

```python
# Export the DataFrame as a CSV file
output_data_file = "../Data/MergedDF.csv"

# Export the City_Data into a CSV.
merged_df.to_csv(output_data_file, index=False)
```
