```python
import pandas as pd
```

```python
file = '../Data/Clean_MergedDF.csv'
df = pd.read_csv(file)

print(df.shape)
df.head()
```

```python
df2 = df.copy()

df2.loc[df2['Median House Value'].between(0, 100000, 'both'), 'house_value_group'] = '<= $100k'
df2.loc[df2['Median House Value'].between(100000, 200000, 'right'), 'house_value_group'] = '$100k - $200k'
df2.loc[df2['Median House Value'].between(200000, 300000, 'right'), 'house_value_group'] = '$200k - $300k'
df2.loc[df2['Median House Value'].between(300000, 400000, 'right'), 'house_value_group'] = '$300k - $400k'
df2.loc[df2['Median House Value'].between(400000, 1000000000, 'right'), 'house_value_group'] = '>$400k'

df2.head()
```

```python
# Export the DataFrame as a CSV file
output_data_file = "../Data/DF_for_Visualizations.csv"

# Export the City_Data into a CSV.
df2.to_csv(output_data_file, index=False)
```
